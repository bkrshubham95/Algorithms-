```python
class complex_:

    def __init__(self, re=0, im=0):
        self.re = re
        self.im =im

    def get_re(self):
        return self.re

    def get_im(self):
        return self.im


    def __str__(self):
        g = lambda x: "+" if x>=0 else ""
        return f"({self.re}{g(self.im)}{self.im}i)"

    def __mul__(self, other):
        new_re= ((self.get_re())*(other.get_re())) + ((self.get_im())*(other.get_im()))
        new_im = ((self.get_re())*(other.get_im())) + ((other.get_re())*(self.get_im()))
        return complex_(new_re, new_im)

    def cmul(self, other):
        new_re= ((self.get_re())*(other.get_re())) + ((self.get_im())*(other.get_im()))
        new_im = ((self.get_re())*(other.get_im())) + ((other.get_re())*(self.get_im()))
        return complex_(new_re, new_im)


if __name__ =="__main__":
        w = complex_(1,-3)
        x = complex_(-1,3)
        y = complex_(1,3)
        z = complex_(-1,-3)

        print(w * x)
        print(w.cmul(x))
        print(w*x*y*z)
        print(w.cmul(x).cmul(y).cmul(z))

```

For this program  idea is to overload the __mul__ default function to perform multiplication for the complex numbers. I have used the complex class already defined in the practice problem . In the main function I create two complex number object and when I do a multiplication of the two complex numbers using Asterisk(*) I get product of the two complex number as the function call goes to __mul__ which I have overloaded to take two  number as input and and return product .
